```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the processed dataset (final dataset from part 2, step 8)
df = pd.read_csv("processed_data.csv")

# Convert 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Preview the dataset
df.head()

```

```python
# Set date as index for time series visualization
df.set_index('date', inplace=True)

# Plot temperatures over time
df['avgTemp'].plot(figsize=(12, 6), title="Daily Average Temperature in 2022", ylabel="Temperature (°C)", xlabel="Date", color="blue")

# Show plot
plt.show()
```

```python
# Count trips per day
trip_counts = df.groupby('date').size().reset_index(name='trip_count')

# Merge trip count data with temperature data
df_merged = df.merge(trip_counts, on='date', how='left')

# Preview merged dataframe
df_merged.head()
```

```python
fig, ax1 = plt.subplots(figsize=(12, 6))

# First y-axis for temperature
ax1.set_xlabel("Date")
ax1.set_ylabel("Temperature (°C)", color="blue")
ax1.plot(df_merged.index, df_merged["avgTemp"], color="blue", label="Temperature")
ax1.tick_params(axis="y", labelcolor="blue")

# Second y-axis for bike trips
ax2 = ax1.twinx()
ax2.set_ylabel("Number of Trips", color="red")
ax2.plot(df_merged.index, df_merged["trip_count"], color="red", label="Bike Trips")
ax2.tick_params(axis="y", labelcolor="red")

# Title and Grid
plt.title("Daily Bike Trips & Temperature (2022)")
fig.tight_layout()
plt.show()
```

In this code, I used Matplotlib to create a dual-axis line chart. The first y-axis (ax1) represents temperature, while the second y-axis (ax2) represents bike trips. By using twinx(), I was able to overlay two different scales on the same x-axis (date). The blue line represents temperature trends, and the red line represents the number of bike trips per day. This visualization helps identify correlations between temperature and bike usage over time.
